```python
#A close match is defined as when either of the following is satisfied:
  #(i)   the match is ""Tied"" (captured in Column Q) 
  #(ii)  Balls remaining at the end of the match (Column V) < =  3
 #(iii) chasing team wins (Column Q) with  <= 1 Wickets remaining (Column R and S)  
 #(iv)  chasing team loses (Column Q) by <= 5 runs. (Column R and S)"
#Break out year by year how many close matches were played

df_mediumans1 = xlipl[
                  (xlipl['Winning_Team'] == 'Match Tied')
                | (xlipl['Balls_Remaining'] <= 3 )
                | ((xlipl['Win_Type']=='run') & (xlipl['Winning_Margin']<=5)) 
                | ((xlipl['Win_Type']=='wicket') & (xlipl['Winning_Margin']<=1))
                ]
```

```python
def percentage_won(a= None,b =None):
    if a == 0 or a == None or b == None:
        raise "Value can not be zero!!"
    return (b/a) * 100
```

```python
#"For the 4 teams listed, determine which team has won the most percentage of their matches batting first across all IPLs. To do this, compute for each team: 
#(i)   A=total number of matches won by the team 
#{ii)  B=how many matches they won batting first
#Percent won batting first = (B / A) *100"
medium2 = {'Chennai': 0, 'Hyderabad':0, 'Mumbai':0, 'Kolkata':0}
```

```python
teams = ["Chennai","Hyderabad","Mumbai","Kolkata"]
for team_name in teams:
    ans_b = xlipl[(xlipl['Team_Batting_First']==team_name) &
                (xlipl['Winner']==team_name)].Match_Number.count()
    ans_a=  xlipl[xlipl['Winner']==team_name].Match_Number.count()
    medium2[team_name] = percentage_won(b=ans_b, a=ans_a)
```

```python
medium2
```

```python
medium2final = pd.DataFrame.from_dict(medium2, orient='index')
medium2final
```
