```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
xlipl=pd.read_excel("All_IPL_Match_Scores_easy.xlsx")
xlipl
```

```python
score_avg = xlipl.groupby('Year')['Bat_First_Runs_Scored'].mean()
```

```python
plt.plot(score_avg.index, score_avg.values,  linestyle='-.', color='green')
plt.title('Average First Inning Scores per IPL Season')
plt.xlabel('Year')
plt.ylabel('Bat_First_Runs_Scored')
plt.grid()
plt.show()
```

```python
match_counts = xlipl.groupby('Winner').Match_Number.count()
```

```python
plt.figure(figsize=(13, 5))
plt.bar(match_counts.index, match_counts.values, color='red', edgecolor='black')
plt.title('Number of Matches Won by Each IPL Team')
plt.xlabel('Team Name')
plt.ylabel('Number of Wins')
plt.legend()
plt.show()
```

```python
wintype_counts = xlipl['Win_Type'].value_counts()
```

```python
plt.pie(wintype_counts, labels=wintype_counts.index, autopct='%1.1f%%', startangle=50, colors=['lightpink', 'green'])
plt.title('Distribution of Wins by Margin Type (Runs vs Wickets)')
plt.axis('equal')
plt.show()
```
